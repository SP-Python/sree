```python
# Read the Language Dashboard Report downloaded from GA for the whole year suggested in requisition form.

lang_dash = pd.read_excel(open('Analytics All Web Site Data Language_Dashboard_N 20160801-20170731.xlsx','rb'), sheetname='Dataset1')
lang_dash1 = lang_dash.dropna()
lang_dash1
```

```python
# including new calculations in the data, [ Pageviews, users, bounces, new users] /12 to get the average for the whole year.
# Bounce Rate, just keep as is.
# include new calc metrics %nvs, %of language, P/V.
# Visits - Sessions (Visits is now sessions GA)
# Unique Visitors - Users (Unique Visitos is now Users in GA)

lang_dash2 = (lang_dash1.groupby(['Language']).agg(({"Pageviews": "sum","Sessions": "sum","Users": "sum","Bounces": "sum","New Users": "sum","Bounce Rate" : "mean"}))/12).sort_values('Pageviews',ascending=False)
lang_dash2['Bounce Rate'] = (lang_dash2['Bounce Rate']*12*100).round(2)
lang_dash2['Pageviews'] = (lang_dash2['Pageviews']).astype(np.int64)
lang_dash2['New Users'] = (lang_dash2['New Users']).astype(np.int64)
lang_dash2['Users'] = (lang_dash2['Users']).astype(np.int64)
lang_dash2['Bounces'] = (lang_dash2['Bounces']).astype(np.int64)
lang_dash2['%nvs'] = ((lang_dash2['New Users'] / lang_dash2['New Users'].sum())*100).round(2)
lang_dash2['% Of Language'] = ((lang_dash2['Users'] / lang_dash2['Users'].sum())*100).round(2)
lang_dash2['P/V'] = (lang_dash2['Pageviews'] / lang_dash2['Sessions']).round(2)

lang_dash3 = lang_dash2[['Pageviews','Users','Bounces','Bounce Rate','New Users','%nvs','% Of Language','P/V']]
lang_dash3.reset_index(level=0, inplace=True)
lang_dash3
```

```python
### HOW TO PREPARE SHIFT DASHBOARD ###
```

```python
# 4. Take the average monthly UVs of the whole year. Take the month approximately matching with the average monthly UVs and 
# that will be used to create the Shift Dashboard.
# Return rows in the DF which is closest to the Average Value of UV.

average_user = round(Uniq_vis1['Users'].mean(),0)
Uniq_vis1['Diff'] = average_user - Uniq_vis1['Users']

Min = Uniq_vis1['Users'] <= average_user
Max = Uniq_vis1['Users'] >= average_user
idx_Min = Uniq_vis1.ix[Min, 'Users'].idxmax()
idx_Max = Uniq_vis1.ix[Max, 'Users'].idxmin()
Uniq_vis1.ix[idx_Min:idx_Max, ['Month of Year','Users','Diff']]

# The one which is extremely closest is Jan 2017
```

```python
shift = pd.read_excel(open('Analytics All Web Site Data Unique_Visitors_Report 20170101-20170131 (2).xlsx','rb'), sheetname='Dataset1')
shift1 = shift.dropna()

shift1['Date'] = shift1['Date'].astype(str).apply(lambda x: pd.to_datetime(x, format='%Y%m%d'))
shift1['day_of_week'] = shift1['Date'].dt.weekday_name
shift1
```

```python
# Now we need to split the Data frame to WeekDays DF and Week End DF.

# 1. We need to take the average UVs of the Weekdays, then try to take the closes match of the weekday average with one of the 
# weekday day and then pull out the hourly UVs data for that Day from GA.

# 2. We need to take the average UVs of the Weekends., then try to take the closes match of the weekend average with one of the 
# weekend day and then pull out the hourly UVs data for that Day from GA.

# With #1 & #2 that we can end up calculating the  (Daily) Weekday average and (Daily) Weekend average

```

```python

# 2. We need to take the average UVs of the Weekends., then try to take the closes match of the weekend average with one of the 
# weekend day and then pull out the hourly UVs data for that Day from GA.

valuelist = ['Saturday','Sunday']
weekend_data = shift1[shift1.day_of_week.isin(valuelist)]
weekend_avg = round(weekend_data['Users'].sum()/weekend_data['Users'].count(),0)

Min = weekend_data['Users'] <= weekend_avg
Max = weekend_data['Users'] >= weekend_avg
idx_Min = weekend_data.ix[Min, 'Users'].idxmax()
idx_Max = weekend_data.ix[Max, 'Users'].idxmin()
weekend_data.ix[idx_Min:idx_Max, ['Date','Users','day_of_week']]

## we can take Jan 7th for the Weekend hourly data
```

```python
# 1. We need to take the average UVs of the Weekdays, then try to take the closes match of the weekday average with one of the 
# weekday day and then pull out the hourly UVs data for that Day from GA.

valuelist1 = ['Monday','Tuesday','Wednesday','Thursday','Friday']
weekday_data = shift1[shift1.day_of_week.isin(valuelist1)]
weekday_avg = round(weekday_data['Users'].sum()/weekday_data['Users'].count(),0)

Min = weekday_data['Users'] <= weekday_avg
Max = weekday_data['Users'] >= weekday_avg
idx_Min = weekday_data.ix[Min, 'Users'].idxmax()
idx_Max = weekday_data.ix[Max, 'Users'].idxmin()
weekday_data.ix[idx_Min:idx_Max, ['Date','Users','day_of_week']]

## we can take Jan 5th for the Weekeday hourly data
```

```python
# download the weekend hourly data from Google Analytics and Loading the Weekend hourly data.

weekend_hourly = pd.read_excel(open('Analytics All Web Site Data Unique_Visitors_Report 20170107-20170107.xlsx','rb'), sheetname='Dataset1')
weekend_hourly = weekend_hourly.dropna()
weekend_hourly.columns = ['Month of Year','Hour', '(Daily) Weekend average','Unique Pageviews']
weekend_hourly2 = weekend_hourly[['Hour','(Daily) Weekend average']]
weekend_hourly2
```

```python
# download the weekday hourly data from Google Analytics and Loading the Weekend hourly data.

weekday_hourly = pd.read_excel(open('Analytics All Web Site Data Unique_Visitors_Report 20170105-20170105.xlsx','rb'), sheetname='Dataset1')
weekday_hourly1 = weekday_hourly.dropna()
weekday_hourly1.columns = ['Month of Year','Hour', '(Daily) Weekday average','Unique Pageviews']
weekday_hourly2 = weekday_hourly1[['Hour','(Daily) Weekday average']]
weekday_hourly2
```

```python
shift_data = pd.merge(weekday_hourly2, weekend_hourly2,on='Hour', how = 'left')
shift_data.columns = ['Hour','(Daily) Weekday average','(Daily) Weekend average']
shift_data['Total'] = (shift_data['(Daily) Weekday average'] * 5) + (shift_data['(Daily) Weekend average'] * 2).astype(np.int64)
shift_data['Total Average'] = (shift_data['Total'] / 7).astype(np.int64)
shift_data['8*5 Shift'] = (shift_data['(Daily) Weekday average'] * 5).astype(np.int64)
shift_data['12*5 Shift'] = (shift_data['(Daily) Weekday average'] * 5).astype(np.int64)
shift_data['8*7 Shift'] = shift_data['Total'].astype(np.int64)
shift_data['12*7 Shift'] = shift_data['Total'].astype(np.int64)
shift_data['18*5 Shift'] = (shift_data['(Daily) Weekday average'] * 5).astype(np.int64)
shift_data['18*7 Shift'] = shift_data['Total'].astype(np.int64)
shift_data.loc['Sum Total'] = shift_data.sum().astype(np.int64)
shift_data.loc['Percentage of Traffic'] = ((shift_data.sum()/shift_data['Total'].sum())*100).astype(np.int64)
shift_data
```

```python
# Shift Dashboard is saved in the Template. 

book = load_workbook('ProformaReport_Template.xlsx')
writer = pd.ExcelWriter('ProformaReport_Template.xlsx', engine='openpyxl') 
writer.book = book
writer.sheets = dict((ws.title, ws) for ws in book.worksheets)

shift_data.to_excel(writer, "Shift_Dashboard")

writer.save()
```

```python
# Download the pages report for the particular month which has been requested in the proforma requisiton form. 

pages_report = pd.read_excel(open('Analytics All Web Site Data LandingPagesReport 20170701-20170731.xlsx','rb'), sheetname='Dataset1')
pages_report1 = pages_report.dropna()
```

```python
pages_report1['In Scope Language'] = 'English'
x = Lang_Dash_final.ix[(Lang_Dash_final.Language == 'en-us').idxmax(), '% Of Language']
pages_report1['In Scope Language %'] = x
pages_report1['In Scope Visitors'] = ((pages_report1['Users'] * pages_report1['In Scope Language %'])/100).astype(np.int64)
pages_report1['e-Commerce Conversion'] = (pages_report1['Transactions']/ pages_report1['Sessions'])*100
pages_report1['Page Conversion'] = (pages_report1['Transactions']/ pages_report1['Users'])*100
pages_report1['AOV'] = (pages_report1['Revenue'] / pages_report1['Transactions']).astype(np.float32)
pages_report1['RPV'] = (pages_report1['Revenue'] / pages_report1['Users']).astype(np.float32)
pages_report1['Exit %'] = (pages_report1['Exits'] / pages_report1['Pageviews'])
pages_report1['Time Threshold (In seconds)'] = 0.20
```
